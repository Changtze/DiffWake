```python
@dataclass
class YawConstraints:
    gamma_min: float  # minimum allowable yaw angle
    gamma_max: float  # maximum allowable yaw angle

    def mins(self, dtype) -> jax.Array:
        return jnp.deg2rad(jnp.array([self.gamma_min], dtype=dtype))

    def maxs(self, dtype) -> jax.Array:
        return jnp.deg2rad(jnp.array([self.gamma_max], dtype=dtype))

def _latin_hypercube_unit(key, M: int, D: int, dtype) -> jax.Array:
    base = jnp.arange(M, dtype=jnp.int32)
    keys = jax.random.split(key, D + 1)
    k_jit, k_perms = keys[0], keys[1:]
    jitter = jax.random.uniform(k_jit, shape=(D, M), dtype=dtype)

    def _perm_one(k):
        return jax.random.permutation(k, base)

    perms = jax.vmap(_perm_one)(k_perms)
    U = (perms.astype(dtype) + jitter) / dtype(M)
    return U.T


def sample_initial_yaw_lhs(key,
                           M: int,
                           N: int,
                           constraints: YawConstraints,
                           dtype) -> jax.Array:
    U = _latin_hypercube_unit(key, M, 1 * N, dtype)
    U = U.reshape(M, N, 1)
    mins = constraints.mins(dtype)
    maxs = constraints.maxs(dtype)
    gamma = mins + (maxs - mins) * U[:, :, 0]
    return gamma

DTYPE = set_dtype()

yaw_constraints = YawConstraints(0, 25.0)
key1 = jax.random.PRNGKey(42)
key2 = jax.random.PRNGKey(43)
key3 = jax.random.PRNGKey(44)

a = sample_initial_yaw_lhs(key1, 1, 3, yaw_constraints, jnp.float32)
b = sample_initial_yaw_lhs(key2, 1, 3, yaw_constraints, jnp.float32)
c = sample_initial_yaw_lhs(key3, 1, 3, yaw_constraints, jnp.float32)

gamma_min = jnp.deg2rad(0.0).item()
gamma_max = jnp.deg2rad(25.0).item()

N = 3
domain = {f'x{i}': bayex.domain.Real(gamma_min, gamma_max) for i in range(N)}
optimizer = bayex.Optimizer(domain=domain, maximize=False, acq='EI')

# Define prior evaluations to initialise the GP
num_init_evals = 10  # number of initial evaluations
subkeys = jax.random.split(key1, num_init_evals)

# Initial array of evaluations in the shape (N, num_init_evals, 1)
init_evals = sample_initial_yaw_lhs(key1, num_init_evals, N, yaw_constraints, DTYPE)


def f(x):
    return -jnp.sum(x)**2

params = {f'x{i}': init_evals[:, i].astype(DTYPE) for i in range(N)}
ys = jax.vmap(f)(init_evals).astype(DTYPE)
opt_state = optimizer.init(ys, params)

# Number of optimization steps you want to run
n_steps = 15

# Ensure we have a working key for the loop
key = jax.random.PRNGKey(100)

for step in range(n_steps):
    # 1. Split the JAX PRNG key for the random sampling process
    key, subkey = jax.random.split(key)

    # 2. Ask the optimizer for the next best parameters to try.
    # This returns a dictionary (e.g., {'x0': Array, 'x1': Array, 'x2': Array})
    new_params = optimizer.sample(subkey, opt_state)

    # 3. Pack those dictionary values back into a 1D JAX array of shape (N,)
    # so it matches what your objective function f(x) expects.
    x_new = jnp.array([new_params[f'x{i}'] for i in range(N)])

    # 4. Evaluate your objective function to get the scalar 'y' result
    y_new = f(x_new)

    # 5. Update the optimizer state with the new observation.
    # We pass in the scalar y_new and the dictionary new_params.
    opt_state = optimizer.fit(opt_state, y_new, new_params)

    # Print the progress
    print(f"Step {step + 1:>2} | New Eval y: {y_new:+.4f} | Current Best: {opt_state.best_score:+.4f}")

# After the loop finishes, you can extract the best parameters found overall:
best_x_array = jnp.array([opt_state.best_params[f'x{i}'] for i in range(N)])
print(f"\nOptimization Complete.")
print(f"Global Best Score: {opt_state.best_score:.4f}")
print(f"Best Parameters: {best_x_array}")
```

## Numerical precision tests
Physical simulations require accuracy but there can be a big runtime tradeoff when switching between FP16, FP32 and FP64 precision.
Let's use 4 cases:
1. A row of 3 turbines at x = [0.0, 400.0, 800.0], unyawed
2. A row of 3 turbines at x = [0.0, 400.0, 800.0], yawed at [20.0, 10.0, 0.0]
3. Horns Rev 1, unyawed
4. Horns Rev 1, all yawed at 10 degrees except the most downstream row, which is left unyawed.

All cases have the same wind speed, wind direction and turbulence intensity.

```python
wake_model = "gauss"
data_dir = Path("../data/horn")
farm_cfg_path = data_dir / f"gch.yaml"
turbine_cfg_path = data_dir / "vestas_v802MW.yaml"


# Set FP64
from diffwake.diffwake_jax.util import set_precision
DTYPE = set_precision("fp64")
print(f"Current floating point precision: {DTYPE}")

# Base configuration (unyawed)
cfg = load_input(farm_cfg_path, turbine_cfg_path)
ws_test = 13.5
wd_test = 270.0
ti_val_test = 0.06
layout_x = jnp.array([0.0, 400.0, 800.0], dtype=DTYPE)
layout_y = jnp.array([0.0, 0.0, 0.0], dtype=DTYPE)
wind_speed = jnp.array([ws_test], dtype=DTYPE)
wind_dir = jnp.array([jnp.deg2rad(wd_test)], dtype=DTYPE) # Wind from West
ti = jnp.array([ti_val_test], dtype=DTYPE)

def run_configuration(base_config: Config,
                      ws: jnp.ndarray,
                      wd: jnp.ndarray,
                      ti_val: jnp.ndarray,
                      layout: Tuple[jnp.ndarray, jnp.ndarray] = (None, None),
                      yaw_angles: jnp.ndarray = None,
                      dtype = jnp.float64):
    if None in layout:
        test_config = base_config.set(
            wind_speeds=ws,
            wind_directions=wd,
            turbulence_intensities=ti_val,
        )
    else:
        test_config = base_config.set(
            layout_x=layout[0],
            layout_y=layout[1],
            wind_speeds=ws,
            wind_directions=wd,
            turbulence_intensities=ti_val,
        )

    test_state = create_state(test_config)
    test_state = jax.tree_util.tree_map(
        lambda x: x.astype(dtype) if hasattr(x, "dtype") and jnp.issubdtype(x.dtype, jnp.floating) else x, test_state
    )

    if yaw_angles is not None:
        test_state_yawed = alter_yaw_angles(yaw_angles, test_state)
        time_start = time.time()
        test_result = simulate(test_state_yawed)
        jax.tree_util.tree_map(lambda x: x.block_until_ready(), test_result)
        time_end = time.time()
        tp = power_fn(power_thrust_table=test_state_yawed.farm.power_thrust_table,
                                      velocities=test_result.u_sorted,
                                      air_density=test_state_yawed.flow.air_density,
                                      yaw_angles=test_state_yawed.farm.yaw_angles)

    else:
        time_start = time.time()
        test_result = simulate(test_state)
        jax.tree_util.tree_map(lambda x: x.block_until_ready(), test_result)
        time_end = time.time()
        tp = power_fn(power_thrust_table=test_state.farm.power_thrust_table,
                                      velocities=test_result.u_sorted,
                                      air_density=test_state.flow.air_density,
                                      yaw_angles=test_state.farm.yaw_angles)


    # Watts
    farm_power = jnp.sum(tp)

    return test_result, tp, farm_power, time_end - time_start
```

### Run FP64 tests for all 4 configurations

Let's get an average runtime for each configuration in the unyawed and yawed states. If we execute run_configuration for 4 different cases 50 times, we will be running 200 simulations.

```python
# ENABLE X64 AND VERIFY DTYPE IS FP64

print(f"x64 enabled: {jax.config.x64_enabled}")

def run_all_cases(config, dtype=jnp.float64):
    # Cases 1 and 2
    case1 = run_configuration(config, wind_speed, wind_dir, ti, layout=(layout_x, layout_y), dtype=dtype)
    case2 = run_configuration(config, wind_speed, wind_dir, ti, layout=(layout_x, layout_y), yaw_angles=jnp.deg2rad(jnp.array([[20.0, 10.0, 0.0]])), dtype=dtype)

    # Cases 3 and 4
    hr1_yaw = jnp.deg2rad(jnp.array([jnp.concatenate([jnp.full(70, 10.0), jnp.zeros(10)])]))
    case3 = run_configuration(config, wind_speed, wind_dir, ti, dtype=dtype)
    case4 = run_configuration(config, wind_speed, wind_dir, ti, yaw_angles=hr1_yaw, dtype=dtype)

    return case1, case2, case3, case4

```

```python
# Get the average runtime for all cases over N_runs runs using FP64
N_runs = 50

# Tracking runtimes
rt_1 = []
rt_2 = []
rt_3 = []
rt_4 = []

for i in range(N_runs):
    c1_fp64, c2_fp64, c3_fp64, c4_fp64 = run_all_cases(config=cfg, dtype=jnp.float64)

    # Append all runtimes and get the average
    rt_1.append(c1_fp64[3])
    rt_2.append(c2_fp64[3])
    rt_3.append(c3_fp64[3])
    rt_4.append(c4_fp64[3])
```
